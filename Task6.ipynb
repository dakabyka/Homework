# Решение СЛАУ методом Гаусса

```python
import numpy as np
import random
import scipy.linalg as ln
```

```python
p = 20
k = np.zeros((p, p + 1))
for i in range(p):
    k[i, i] = np.random.randint(100)
    k[i, p] = np.random.randint(100)
for i in range(p - 1):
    k[i + 1, i] = np.random.randint(100)
    k[i, i+1] = np.random.randint(100)

    
```

```python
def slau(k):
    p = len(k)
    x = np.zeros(p)
    for i in range(p - 1):
        k[i+1, 0:] = k[i+1, 0:] - k[i,0:]/k[i,i]*k[i+1,i]
    for i in range(p-1, 0, -1):
        k[i-1,0:] = k[i-1,0:] - k[i,0:]*k[i-1,i]/k[i,i]
    for i in range(p):
        x[i] = (k[i,p]/k[i,i])
    return x
```
