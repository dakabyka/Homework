$$x''+w^2x = kx^3$$

## Решение явным методом Эйлера

```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

Пусть x' = v

```python
def f(x, v, c):
    return -c[0]**2*x-c[1]*v+c[2]*x**3
```

```python
print('Параметры')
c = list(map(float, input().split(',')))
print('Начальные условия', 'x_0=', sep ='\n', end ='')
x = [float(input())]
print('v_0=', end ='')
v = [float(input())]
h = np.linspace(0.0, 10*2*pi/c[0], 10**(6))
w = h[1] - h[0]
for t_i in h[1:]:
    v.append(v[-1] + w * f(x[-1], v[-1], c))
    x.append(x[-1] + w * v[-2])
plt.plot(h, x, '-')
```

## Решение методом Ранге-Кутта второго порядка

```python
print('Параметры')
c = list(map(float, input().split(',')))
print('Начальные условия', 'x_0=', sep ='\n', end ='')
x = [float(input())]
print('v_0=', end ='')
v = [float(input())]
h = np.linspace(0.0, 10*2*pi/c[0], 10**(6))
w = h[1] - h[0]
for t_i in h[1:]:
    v.append(v[-1] + w * f(0.5 * w + x[-1], v[-1] + 0.5 * w * f(x[-1] , v[-1], c), c))
    x.append(x[-1] + w * v[-2])
plt.plot(h, x, '-')
```
